````python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread("bild.jpg")
````

````python
i = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,10))
plt.imshow(i)
plt.show()
````

````python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(20,10))
plt.imshow(gray, "gray")
plt.show()
````

````python
classifier = cv2.CascadeClassifier("haarcascades/haarcascade_frontalface_alt2.xml")
````

````python
faces = classifier.detectMultiScale(gray, minNeighbors=10)
````

````python
print(faces)
````

````python
c = img.copy()
for face in faces:
    x, y, w, h = face
    cv2.rectangle(c, (x, y), (x + w, y + h), (0, 255, 0), 10)
    print(face)

i = cv2.cvtColor(c, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,10))
plt.imshow(i)
plt.show()
````
